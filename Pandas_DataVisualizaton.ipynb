### Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
timeseries = pd.Series(np.random.randint(1,10,size=(100))) # in series data must be one dimensional
timeseries = timeseries.cumsum()
timeseries.plot()
```

```python
timeseries = pd.DataFrame(np.random.randint(1,10,size=(10,10))) 
timeseries = timeseries.cumsum()
plt.figure()
timeseries.plot()
```
